###first_member and second_member

function first_member first determines if there are multiple members by determining the number of commas in the "List of Members" column. After that it determines if there's 1 or 3 commas in the entry. If there's one, then just take the name before the single comma. If there's three, then this means that the name's format is SURNAME, FIRST NAME. So, the name in between the first and second comma is the first person's first name and whatever's before the first comma is the first person's surname



```python
def first_member(abc):
  if(abc.count(',')==0):
    if(abc.count('/')>0):
      abc.replace('/', ' ')
    if(abc.count('\n')>0):
      return abc[0:abc.find('\n')]
    else:
      return abc
  elif(abc.count(',')==1):
    return abc[0:abc.find(',')]
  elif(abc.count(',')==3):
    return abc[abc.find(',')+1:abc.find(',', abc.find(',')+1)]+abc[0:abc.find(',')]

def second_member(abc):
  if(abc.count(',')==0):
    if(abc.count('\n')>0):
      return abc[abc.find('\n')+1:abc.find('\n', abc.find('\n')+1)]
    else:
      return 
  elif(abc.count(',')==1):
    return abc[abc.find(',')+1:]#abc.find(',', abc.find(',')+1)]
  elif(abc.count(',')==3):
    return abc[abc.rfind(',')+1:]+abc[abc.find(',', abc.find(',')+3):abc.rfind(',')]
```

function second_member almost has the same logic as the first function. The difference are the indices used to extract the names.
